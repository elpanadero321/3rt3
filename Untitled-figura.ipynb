```python
import io
import numpy as np
import pandas as pd
import requests
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp, cumulative_trapezoid
from scipy.optimize import minimize

C_LIGHT = 299792.458

PANTHEON_DATA_URL = "https://raw.githubusercontent.com/PantheonPlusSH0ES/DataRelease/main/Pantheon+_Data/4_DISTANCES_AND_COVAR/Pantheon+SH0ES.dat"
PANTHEON_COV_URL = "https://raw.githubusercontent.com/PantheonPlusSH0ES/DataRelease/main/Pantheon+_Data/4_DISTANCES_AND_COVAR/Pantheon+SH0ES_STAT+SYS.cov"

BAO_BLOCKS = [
    {"z": 0.295, "kind": "DV", "mean": 7.93, "sigma": 0.15},
    {"z": 0.510, "kind": "DM_DH", "mean": [13.62, 20.98], "sigma": [0.25, 0.61], "rho": -0.445},
    {"z": 0.706, "kind": "DM_DH", "mean": [16.85, 20.08], "sigma": [0.32, 0.60], "rho": -0.420},
    {"z": 0.930, "kind": "DM_DH", "mean": [21.71, 17.88], "sigma": [0.28, 0.35], "rho": -0.389},
    {"z": 1.317, "kind": "DM_DH", "mean": [27.79, 13.82], "sigma": [0.69, 0.42], "rho": -0.444},
    {"z": 1.491, "kind": "DV", "mean": 26.07, "sigma": 0.67},
    {"z": 2.330, "kind": "DM_DH", "mean": [39.71, 8.52], "sigma": [0.94, 0.17], "rho": -0.477},
]

HZ_CC = np.array([
    [0.07, 69.0, 19.6], [0.09, 69.0, 12.0], [0.12, 68.6, 26.2],
    [0.17, 83.0, 8.0], [0.1791, 78.0, 6.2], [0.1993, 78.0, 6.9],
    [0.20, 72.9, 29.6], [0.27, 77.0, 14.0], [0.28, 88.8, 36.6],
    [0.3519, 85.5, 15.7], [0.3802, 86.2, 14.6], [0.40, 95.0, 17.0],
    [0.4004, 79.9, 11.4], [0.4247, 90.4, 12.8], [0.4497, 96.3, 14.4],
    [0.47, 89.0, 49.6], [0.4783, 83.8, 10.2], [0.48, 97.0, 62.0],
    [0.5929, 107.0, 15.5], [0.6797, 95.0, 10.5], [0.75, 98.8, 33.6],
    [0.7812, 96.5, 12.5], [0.8754, 124.5, 17.4], [0.88, 90.0, 40.0],
    [0.90, 117.0, 23.0], [1.037, 133.5, 17.6], [1.30, 168.0, 17.0],
    [1.363, 160.0, 33.8], [1.43, 177.0, 18.0], [1.53, 140.0, 14.0],
    [1.75, 202.0, 40.0], [1.965, 186.5, 50.6],
], dtype=float)


def load_pantheon_plus():
    data_res = requests.get(PANTHEON_DATA_URL, timeout=60)
    cov_res = requests.get(PANTHEON_COV_URL, timeout=60)
    data_res.raise_for_status()
    cov_res.raise_for_status()

    df = pd.read_csv(io.StringIO(data_res.text), sep=r"\s+", comment="#")

    cov_lines = cov_res.text.strip().splitlines()
    n_cov = int(cov_lines[0].strip())
    cov = np.array([float(x) for x in cov_lines[1:]], dtype=float).reshape(n_cov, n_cov)

    mask = (df["IS_CALIBRATOR"] == 0).to_numpy()
    df = df.loc[mask].reset_index(drop=True)
    cov = cov[np.ix_(mask, mask)]

    z = df["zHD"].to_numpy(dtype=float)
    mu = df["MU_SH0ES"].to_numpy(dtype=float)

    good = np.isfinite(z) & np.isfinite(mu) & (z > 0.0)
    z = z[good]
    mu = mu[good]
    cov = cov[np.ix_(good, good)]

    return z, mu, cov


def rhs_dvch(z, y, n, beta):
    om_m, om_l, om_r = y
    om_m = max(om_m, 1e-15)
    om_l = max(om_l, 0.0)
    om_r = max(om_r, 0.0)

    e2 = max(om_m + om_l + om_r, 1e-20)
    e = np.sqrt(e2)

    qtilde = -e * om_l / (1.0 + n * om_l / om_m) * (
        n - (beta / (1.0 + beta * e2)) * ((4.0 * om_r + 3.0 * om_m) / 3.0)
    )

    dom_m = 3.0 * (om_m + qtilde / e) / (1.0 + z)
    dom_l = -3.0 * qtilde / (e * (1.0 + z))
    dom_r = 4.0 * om_r / (1.0 + z)
    return [dom_m, dom_l, dom_r]


def solve_dvch_grid(zmax, H0, Om0, Or0, n, beta, npts=3500):
    OL0 = 1.0 - Om0 - Or0
    if OL0 <= 0.0:
        raise ValueError("Omega_Lambda0 must be positive.")

    z_grid = np.linspace(0.0, zmax, npts)
    sol = solve_ivp(
        lambda z, y: rhs_dvch(z, y, n=n, beta=beta),
        (0.0, zmax),
        [Om0, OL0, Or0],
        t_eval=z_grid,
        rtol=1e-9,
        atol=1e-12,
    )

    if not sol.success:
        raise RuntimeError(sol.message)

    om_m, om_l, om_r = sol.y
    E = np.sqrt(np.maximum(om_m + om_l + om_r, 1e-20))
    return z_grid, E


def lcdm_grid(zmax, Om0, Or0, npts=3500):
    z_grid = np.linspace(0.0, zmax, npts)
    OL0 = 1.0 - Om0 - Or0
    E = np.sqrt(Om0 * (1.0 + z_grid) ** 3 + Or0 * (1.0 + z_grid) ** 4 + OL0)
    return z_grid, E


def distance_grids(z_grid, E_grid, H0):
    invE = 1.0 / E_grid
    chi = (C_LIGHT / H0) * cumulative_trapezoid(invE, z_grid, initial=0.0)
    DH = C_LIGHT / (H0 * E_grid)
    return chi, DH


def mu_from_grid(z_data, z_grid, chi_grid, M_shift):
    chi_at = np.interp(z_data, z_grid, chi_grid)
    dL = (1.0 + z_data) * chi_at
    return 5.0 * np.log10(dL) + 25.0 + M_shift


def pantheon_chi2(z_grid, E_grid, H0, M_shift, z_sn, mu_sn, covinv_sn):
    chi_grid, _ = distance_grids(z_grid, E_grid, H0)
    mu_th = mu_from_grid(z_sn, z_grid, chi_grid, M_shift)
    delta = mu_sn - mu_th
    return float(delta @ covinv_sn @ delta), mu_th


def bao_observables(z_grid, E_grid, H0, rd):
    chi_grid, DH_grid = distance_grids(z_grid, E_grid, H0)
    rows = []

    for block in BAO_BLOCKS:
        z = block["z"]
        DM = np.interp(z, z_grid, chi_grid)
        DH = np.interp(z, z_grid, DH_grid)

        if block["kind"] == "DV":
            DV = (z * DM * DM * DH) ** (1.0 / 3.0)
            pred = DV / rd
            pull = (pred - block["mean"]) / block["sigma"]
            rows.append((z, "DV/rd", block["mean"], pred, pull))
        else:
            pred_dm = DM / rd
            pred_dh = DH / rd
            pull_dm = (pred_dm - block["mean"][0]) / block["sigma"][0]
            pull_dh = (pred_dh - block["mean"][1]) / block["sigma"][1]
            rows.append((z, "DM/rd", block["mean"][0], pred_dm, pull_dm))
            rows.append((z, "DH/rd", block["mean"][1], pred_dh, pull_dh))

    return rows


def bao_chi2(z_grid, E_grid, H0, rd):
    chi_grid, DH_grid = distance_grids(z_grid, E_grid, H0)
    total = 0.0

    for block in BAO_BLOCKS:
        z = block["z"]
        DM = np.interp(z, z_grid, chi_grid)
        DH = np.interp(z, z_grid, DH_grid)

        if block["kind"] == "DV":
            DV = (z * DM * DM * DH) ** (1.0 / 3.0)
            pred = DV / rd
            total += ((pred - block["mean"]) / block["sigma"]) ** 2
        else:
            pred = np.array([DM / rd, DH / rd])
            mean = np.array(block["mean"])
            s1, s2 = block["sigma"]
            rho = block["rho"]
            cov = np.array([
                [s1 * s1, rho * s1 * s2],
                [rho * s1 * s2, s2 * s2],
            ])
            delta = pred - mean
            total += float(delta @ np.linalg.inv(cov) @ delta)

    return total


def hz_chi2(z_grid, E_grid, H0):
    z = HZ_CC[:, 0]
    H_obs = HZ_CC[:, 1]
    H_err = HZ_CC[:, 2]
    H_th = H0 * np.interp(z, z_grid, E_grid)
    return float(np.sum(((H_obs - H_th) / H_err) ** 2)), H_th


def joint_fit():
    z_sn, mu_sn, cov_sn = load_pantheon_plus()
    covinv_sn = np.linalg.inv(cov_sn)

    Or0 = 9e-5
    beta = 1e-4
    zmax = max(np.max(z_sn), np.max(HZ_CC[:, 0]), max(b["z"] for b in BAO_BLOCKS))

    def chi2_dvch(theta):
        H0, Om0, n, rd, M_shift = theta
        if not (50 < H0 < 90 and 0.05 < Om0 < 0.6 and 0 < n < 1 and 120 < rd < 170 and -1 < M_shift < 1):
            return 1e50
        try:
            z_grid, E_grid = solve_dvch_grid(zmax, H0, Om0, Or0, n, beta)
            chi_sn, _ = pantheon_chi2(z_grid, E_grid, H0, M_shift, z_sn, mu_sn, covinv_sn)
            chi_bao = bao_chi2(z_grid, E_grid, H0, rd)
            chi_hz, _ = hz_chi2(z_grid, E_grid, H0)
            return chi_sn + chi_bao + chi_hz
        except Exception:
            return 1e50

    def chi2_lcdm(theta):
        H0, Om0, rd, M_shift = theta
        if not (50 < H0 < 90 and 0.05 < Om0 < 0.6 and 120 < rd < 170 and -1 < M_shift < 1):
            return 1e50
        try:
            z_grid, E_grid = lcdm_grid(zmax, Om0, Or0)
            chi_sn, _ = pantheon_chi2(z_grid, E_grid, H0, M_shift, z_sn, mu_sn, covinv_sn)
            chi_bao = bao_chi2(z_grid, E_grid, H0, rd)
            chi_hz, _ = hz_chi2(z_grid, E_grid, H0)
            return chi_sn + chi_bao + chi_hz
        except Exception:
            return 1e50

    res_dvch = minimize(
        chi2_dvch,
        x0=np.array([73.0, 0.30, 0.20, 147.0, 0.0]),
        method="Nelder-Mead",
        options={"maxiter": 12000, "xatol": 1e-6, "fatol": 1e-6},
    )

    res_lcdm = minimize(
        chi2_lcdm,
        x0=np.array([73.0, 0.30, 147.0, 0.0]),
        method="Nelder-Mead",
        options={"maxiter": 12000, "xatol": 1e-6, "fatol": 1e-6},
    )

    H0_d, Om0_d, n_d, rd_d, M_d = res_dvch.x
    H0_l, Om0_l, rd_l, M_l = res_lcdm.x

    z_grid_d, E_grid_d = solve_dvch_grid(zmax, H0_d, Om0_d, Or0, n_d, beta)
    z_grid_l, E_grid_l = lcdm_grid(zmax, Om0_l, Or0)

    chi_sn_d, mu_d = pantheon_chi2(z_grid_d, E_grid_d, H0_d, M_d, z_sn, mu_sn, covinv_sn)
    chi_sn_l, mu_l = pantheon_chi2(z_grid_l, E_grid_l, H0_l, M_l, z_sn, mu_sn, covinv_sn)

    chi_hz_d, H_d = hz_chi2(z_grid_d, E_grid_d, H0_d)
    chi_hz_l, H_l = hz_chi2(z_grid_l, E_grid_l, H0_l)

    bao_rows_d = bao_observables(z_grid_d, E_grid_d, H0_d, rd_d)
    bao_rows_l = bao_observables(z_grid_l, E_grid_l, H0_l, rd_l)

    chi2_d = float(res_dvch.fun)
    chi2_l = float(res_lcdm.fun)

    N = len(z_sn) + len(HZ_CC) + sum(1 if b["kind"] == "DV" else 2 for b in BAO_BLOCKS)
    AIC_d = chi2_d + 2 * 5
    AIC_l = chi2_l + 2 * 4
    BIC_d = chi2_d + 5 * np.log(N)
    BIC_l = chi2_l + 4 * np.log(N)

    fig, ax = plt.subplots(2, 2, figsize=(13, 10))
    ax = ax.flatten()

    idx = np.linspace(0, len(z_sn) - 1, min(250, len(z_sn))).astype(int)
    ax[0].scatter(z_sn[idx], mu_sn[idx] - mu_l[idx], s=10, alpha=0.6, label=r"Pantheon+ residuals to $\Lambda$CDM")
    order = np.argsort(z_sn)
    ax[0].plot(z_sn[order], (mu_d - mu_l)[order], color="crimson", lw=2, label="DVCH - LCDM")
    ax[0].axhline(0, color="k", ls="--", lw=1)
    ax[0].set_xlabel("z")
    ax[0].set_ylabel(r"$\Delta\mu$")
    ax[0].set_title("Pantheon+")
    ax[0].legend(fontsize=8)

    ax[1].errorbar(HZ_CC[:, 0], HZ_CC[:, 1], yerr=HZ_CC[:, 2], fmt="o", ms=4, alpha=0.7, label="H(z) data")
    z_plot = np.linspace(0, np.max(HZ_CC[:, 0]), 400)
    ax[1].plot(z_plot, H0_d * np.interp(z_plot, z_grid_d, E_grid_d), color="crimson", lw=2, label="DVCH")
    ax[1].plot(z_plot, H0_l * np.interp(z_plot, z_grid_l, E_grid_l), "k--", lw=2, label=r"$\Lambda$CDM")
    ax[1].set_xlabel("z")
    ax[1].set_ylabel(r"$H(z)$ [km s$^{-1}$ Mpc$^{-1}$]")
    ax[1].set_title("Cosmic chronometers")
    ax[1].legend(fontsize=8)

    x_bao = np.arange(len(bao_rows_d))
    pulls_d = np.array([row[4] for row in bao_rows_d])
    pulls_l = np.array([row[4] for row in bao_rows_l])
    labels = [f"{row[1]}\nz={row[0]:.3f}" for row in bao_rows_d]

    width = 0.38
    ax[2].bar(x_bao - width / 2, pulls_d, width=width, color="crimson", alpha=0.8, label="DVCH")
    ax[2].bar(x_bao + width / 2, pulls_l, width=width, color="gray", alpha=0.8, label=r"$\Lambda$CDM")
    ax[2].axhline(0, color="k", ls="--", lw=1)
    ax[2].axhline(1, color="k", ls=":", lw=0.8)
    ax[2].axhline(-1, color="k", ls=":", lw=0.8)
    ax[2].set_xticks(x_bao)
    ax[2].set_xticklabels(labels, rotation=60, ha="right", fontsize=8)
    ax[2].set_ylabel("pull")
    ax[2].set_title("DESI BAO pulls")
    ax[2].legend(fontsize=8)

    ax[3].axis("off")
    text = (
        "Joint real-data fit\n\n"
        f"DVCH: chi2={chi2_d:.3f}, AIC={AIC_d:.3f}, BIC={BIC_d:.3f}\n"
        f"LCDM: chi2={chi2_l:.3f}, AIC={AIC_l:.3f}, BIC={BIC_l:.3f}\n\n"
        f"Delta chi2 = {chi2_d - chi2_l:.3f}\n"
        f"Delta AIC  = {AIC_d - AIC_l:.3f}\n"
        f"Delta BIC  = {BIC_d - BIC_l:.3f}\n\n"
        f"DVCH best fit:\nH0={H0_d:.3f}, Om0={Om0_d:.3f}, n={n_d:.3f}, rd={rd_d:.3f}\n\n"
        f"LCDM best fit:\nH0={H0_l:.3f}, Om0={Om0_l:.3f}, rd={rd_l:.3f}"
    )
    ax[3].text(0.02, 0.98, text, va="top", fontsize=11)

    fig.suptitle("DVCH vs LCDM: joint statistical test with real data", fontsize=15)
    plt.tight_layout()
    plt.savefig("dvch_joint_real_statistics_fixed.png", dpi=300, bbox_inches="tight")
    plt.show()


if __name__ == "__main__":
    joint_fit()
```
